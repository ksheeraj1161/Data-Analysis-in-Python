# Data visualization with different Charts in Python

Data Visualization is the presentation of data in graphical format. It helps people understand the significance of data by summarizing and presenting huge amount of data in a simple and easy-to-understand format and helps communicate information clearly and effectively.

# Different Types of Charts for Analyzing & Presenting Data
# 1. Histogram :
It represents the frequency of occurrence of specific phenomena which lie within a specific range of values and arranged in consecutive and fixed intervals.

```python
import pandas as pd
import matplotlib.pyplot as plt

#create 2d array
#0 marks for overweight and obesity, 10 marks for underweight, 20 marks for normal
data = [['S001', 'M', 44, 23, 'Normal', 87],
		['S002', 'F', 50, 44, 'Overweight', 94],
		['S003', 'F', 47, 35, 'Obesity', 82],
		['S004', 'M', 40, 39, 'Underweight', 89],
		['S005', 'F', 44, 37, 'Underweight', 91],
		['S006', 'M', 36, 21, 'Normal', 77],
		['S007', 'M', 32, 33, 'Obesity', 55],
		['S008', 'F', 26, 40, 'Normal', 86],
		['S009', 'M', 34, 33, 'Normal', 77],
		['S010', 'M', 36, 35, 'Underweight', 81] ]

#create dataframe
df = pd.DataFrame(data, columns = ['STDID', 'Gender',
									'Test1', 'Test2',
									'BMI', 'Result'] )

#create histogram and show data
df.hist()

plt.show()
```

# 2. Column Chart :
To show a comparison among different attributes, or it can show a comparison of items over time.

```python
# Plot the bar chart for numeric values
df.plot.bar()

# Provide 2 attributes
plt.bar(df['Test1'], df['Test2'])
plt.xlabel("Test1")
plt.ylabel("Test2")
plt.show()
```

# 3. Box plot chart :
It is a graphical representation of statistical data based on the minimum, first quartile, median, third quartile, and maximum. 

```python
# For each numeric attribute of dataframe
df.plot.box()

# individual attribute box plot
plt.boxplot(df['Result'])
plt.show()
```

# 4. Pie Chart :
It shows a static number and how categories represent part of a whole the composition of something. A pie chart represents numbers in percentages, and the total sum of all segments needs to equal 100%.

```python
#pie chart 1
plt.pie(df['Test1'], labels = {"A", "B", "C",
							"D", "E", "F",
							"G", "H", "I", "J"},
							
autopct ='% 1.1f %%', shadow = True)
plt.show()

#pie chart 2
plt.pie(df['Test1'], labels = {"A", "B", "C",
								"D", "E", "F",
								"G", "H", "I", "J"},
								
autopct ='% 1.1f %%', shadow = True)
plt.show()

#pie chart 3
plt.pie(df['Result'], labels = {"A", "B", "C",
							"D", "E", "F",
							"G", "H", "I", "J"},
autopct ='% 1.1f %%', shadow = True)
plt.show()
```

# 5. Scatter plot :
It shows the relationship between two different variables and it can reveal the distribution trends. 

```python
# scatter plot 1
plt.scatter(df['Test1'], df['Test2'])
plt.show()

# scatter plot 2
plt.scatter(df['Result'], df['Test1'])
plt.show()

# scatter plot 3
plt.scatter(df['Result'], df['Test2'])
plt.show()
```
